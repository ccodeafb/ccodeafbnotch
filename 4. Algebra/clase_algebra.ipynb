```python
a = 2
b = 4
c = -7
d = 2

plt.xlim(-10,10)
plt.ylim(-10,10)

plt.arrow(0,0,a,b, width=0.1, color="green")
plt.arrow(a,b,c,d, width=0.1, color="green")
plt.arrow(0,0,a+c,d+b, width=0.1)
```

```python
def imprimir_vectores(a,b,c,d):
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    plt.arrow(0,0,a,b, width=0.1, color="green")
    plt.arrow(a,b,c,d, width=0.1, color="green")
    plt.arrow(0,0,a+c,b+d, width=0.1, color="red")
    plt.grid()
    plt.show()
    
```

```python
interactive(imprimir_vectores, 
            a=widgets.FloatSlider(value=0,min=-5,max=5,step=0.2),
            b=widgets.FloatSlider(value=0,min=-5,max=5,step=0.2),
            c=widgets.FloatSlider(value=0,min=-5,max=5,step=0.2),
            d=widgets.FloatSlider(value=0,min=-5,max=5,step=0.2))
```
