Transformaciones a espacios de otra dimension $f:R^2 \rightarrow R^3$

Para multiplicar matrices se tiene que cumplir que:

$(m \times n)(n \times l) = m \times l$

$n$: espacio de entrada 

$m$: espacio de salida

$l$: cantidad de vectores a transformar

Las matrices que proyectan de $f:R^n \rightarrow R^m$, tendran una dimension de $(m \times n)$
