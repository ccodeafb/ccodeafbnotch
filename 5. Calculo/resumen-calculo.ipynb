```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return ((x**2)*(x**3+2*x**2+2)*np.sin(x)*np.sqrt(x**2))/(x**2 + 1)

x = np.linspace(-10, 10, 100)
y = list(map(f, x))
plt.plot(x, y)
```

```python
def f(x):
    return np.exp(x)

def f2(x):
    return np.exp(-x)

x = np.linspace(-10, 10, 100)
y = list(map(f, x))
plt.plot(x, y)
y2 = list(map(f2, x))
plt.plot(x, y2)
```

```python
def f(x):
    return np.log(x)

x = np.linspace(0.001, 10, 100)
y = list(map(f, x))
plt.plot(x, y)
```

```python
def f(x):
    return np.sin(x)

def f2(x):
    return np.cos(-x)

x = np.linspace(-10, 10, 100)
y = list(map(f, x))
plt.plot(x, y)
y2 = list(map(f2, x))
plt.plot(x, y2)
```

```python
from ipywidgets import interactive, widgets
```

```python
def dibujar_recta(m, b):
    def recta(x):
        return m*x + b
    
    x = np.linspace(-10, 10, 100)
    y = list(map(recta, x))
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.plot(x, y)
    plt.grid()
    
interactive(dibujar_recta,
            m=widgets.FloatSlider(min=-5, max=5),
            b=widgets.FloatSlider(min=-5, max=5))
```

# El valor de la derivada de una funcion en un punto es la pendiente de la recta tangente a la funcion en ese punto

### sympy

```python
from sympy import symbols, diff

x = symbols('x')
```

```python
f = x**4 - x**3 + 2
f
```

```python
f_prima = diff(f, x)
f_prima
```

```python
x0 = -1
m = f_prima.evalf(subs={x: x0})
m
```

```python
y0 = f.evalf(subs={x: x0})
y0
```

```python
def f(x):
    return x**4 - x**3 + 2

x = np.linspace(-3, 3, 100)
y = list(map(f, x))
plt.plot(x, y)
```

```python
def f(x):
    return x**4 - x**3 + 2

def recta_tangente(x):
    return -7*x - 3

x = np.linspace(-3, 3, 100)
y = list(map(f, x))
plt.plot(x, y, label='f')

y2 = list(map(recta_tangente, x))
plt.plot(x, y2, label='recta tangente')

plt.plot([x0, x0], [-10, 10], c='red')
plt.legend()
```
