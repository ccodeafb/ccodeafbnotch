```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return (x**2)*np.sin((x**3*2*x**2*2))
```

```python
x = np.linspace(-5, 5, 50)   #input (eje x) para la funcion en el rango -5 a 5  con 50 valores
```

```python
y = list(map(f, x))   #se pasa la funcion a todos los valores y se listean
```

```python
plt.plot(x,y)
```

```python
from sympy import symbols, diff
x, y, z = symbols('x y z')
diff(x**2 + 3*y, y)
```

```python
f = x**2 - x + 1
x0 = 3
derivada = diff(f, x) #para derivar respecto a x, no a y
derivada
```

```python
ejemplo = diff(f, x).evalf(subs={x:x0})  #calular la derivada respecto a x cuando x tiene el valor 3
```
