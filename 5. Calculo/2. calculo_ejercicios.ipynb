### Pendiente de una recta

Dada la recta que pasa por los puntos $(0, 3)$ y $(4, 6)$. Calcula su pendiente y dibuja esa recta en el plano.

La formula para calcular la pendiente dados dos puntos de una recta es la siguiente:

$m = arctan(\frac{y_2-y_1}{x_2-x_1})$

$m = (∆y/∆x) = (y2-y1)/(x2-x1)$

```python
y_puntos = (3, 6)
x_puntos = (0, 4)
m = (y_puntos[1]-y_puntos[0])/(x_puntos[1]-x_puntos[0])
print(f'La pendiente es {m}')
```

La recta es del estilo $y = m*x + b$ para tener b sabemos que es 3, porque la recta pasa por $(0,3)$

```python
b = 3
x = np.linspace(-2, 7, 10)
y = m*x + b
import pandas as pd

plt.xlabel("Valores X")
plt.ylabel("Valores Y")
plt.grid()
plt.plot([0, 0], [2.8, 3.2], c='red')
plt.plot([4, 4], [5.8, 6.2], c='red')
plt.plot(x,y)

```

### Ejercicio 6


Realizar un gráfico parecido al que se encuentra abajo, entre los límites -3 y 3.

Utilizar la función fill_between de matplotlib (https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.fill_between.html)

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/f/f2/Integral_as_region_under_curve.svg/1920px-Integral_as_region_under_curve.svg.png" style="width:200px;"/>

```python
def g(x):
    return (x**2+6*x)

x = np.linspace(-5,5, 50)
y = list(map(g, x))
plt.xlabel("Valores X")
plt.ylabel("Valores Y")
plt.fill_between(x,y, color='green',alpha=0.3)
plt.plot(x,g(x),color='green')
```

```python
def g(x):
    return (x**2+6*x)

def dibujar(a,b):
    y = list(map(g, np.linspace(a,b, 50)))
    plt.xlabel("Valores X")
    plt.ylabel("Valores Y")
    plt.fill_between(np.linspace(a,b, 50),y, color='green',alpha=0.3)
    plt.plot(np.linspace(a,b, 50),g(np.linspace(a,b, 50)),color='green')

interactive(dibujar, a=widgets.FloatSlider(value=0,min=-20,max=20,step=1),b=widgets.FloatSlider(value=0,min=-20,max=20,step=1))
```
