```python
from sklearn.datasets import load_iris, load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

import matplotlib.pyplot as plt
```

```python
data = load_iris()
```

```python
X = data['data']
```

```python
X.shape
```

```python
pca = PCA(n_components=4)
pca.fit(X)

X_pca = pca.transform(X)
```

```python
X_pca.shape
```

```python
plt.scatter(X_pca[:, 0],
            X_pca[:, 1],
            c=data['target'],
            s=5)
```

```python
pve = []
for i in range(4):
    pve.append(sum(pca.explained_variance_ratio_[:i+1]))
plt.plot(range(1, 5), pve)
print('eje y: % varianza explicada')
```

```python
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X)
```

```python
plt.scatter(X_tsne[:, 0],
            X_tsne[:, 1],
            c=data['target'],
            s=5)
```
