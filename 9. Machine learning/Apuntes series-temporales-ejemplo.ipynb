```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
serie = np.linspace(0, 30, 100)
serie = np.sin(0.4 + serie + 0.5*np.random.random(100))
serie
```

```python
plt.plot(serie)
```

```python
WINDOW = 5
LOOKAHEAD = 3 
```

```python
X = []
y = []

for i in range(len(serie) - WINDOW - LOOKAHEAD + 1):
    X.append(serie[i:i+WINDOW])
    y.append(serie[i+WINDOW+LOOKAHEAD-1])
```

```python
X = np.array(X)
y = np.array(y)
```
