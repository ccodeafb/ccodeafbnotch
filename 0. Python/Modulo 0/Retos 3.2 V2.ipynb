```python
#Retos bucles

#Reto 1
num = [1.00, 2.00, 3.00, 4.00, 5.00, 6.00, 7.00, 8.00, 9.00, 10.00]
try:
    print("Todos del 1 al 10:")
    print("  For:")
    for i in range(10):
        print(num[i])
    print("  While:")
    i = 0
    while(i<10):
        print(num[i])
        i = i + 1
    print("Pares:")
    print("  For:")
    for i in range(1, 10, 2):
        print(num[i])
    print("  While:")
    i = 1
    while(i<10):
        print(num[i])
        i = i + 2
    print("Impares divisibles por 3:")
    print("  For:")
    for i in range(1, 10, 1):
        if i % 3 == 0: 
            print(num[i-1])
    print("  While:")
    i = 1
    while(i<10):
        i = i + 1
        if i % 3 == 0:  
            print(num[i-1])
except: print("Ha habido un problema")
finally: print("Secuencia completada")
```
