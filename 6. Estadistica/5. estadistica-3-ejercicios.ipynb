### Reto 1

Obtén la tabla de Z-Scores (desde 0 hasta 3), calculando las integrales correspondientes con scipy y almacenando los datos en un DataFrame

```python
#creamos funcion de distribucion normal estandariada

def f(x):
    pi = 3.14159265358979323846
    e = 2.7182818284590452353602874713527
    sigma = 1
    media = 0
    return (1/(sigma*(2*pi)**0.5))*e**((-0.5)*((x-media)/sigma)**2) 


#declaramos z-scores

z = list(np.linspace(0,34,35)/10)

#calculamos la probabilidad acumulada con la integral - Buscar para ahcerlo tb con un arange de numpy

def integral(u):
    area = []
    for ui in u:
        area.append(quad(f,0,ui))
    area2 = pd.DataFrame(area)
    del(area2[1])
    return area2

#guardamos los resultados en un dataframe

df = integral(z)
df["Z"]=z
df.columns = ["Probabilidad","Z-Score"]
df.set_index("Z-Score", inplace=True)
df_final = df.transpose()
df_final

```

### Reto 3

- Genera 1000 numeros aleatorios de una distribucion normal.


```python
numeros = []

for i in range(1,1001):
    numeros.append(np.random.normal())  
  
numeros
```

- Calcula la distribucion de probabilidad acumulada.

```python
acumulada = [0]*1000
for i,j in enumerate(numeros):
    acumulada[i] = (numeros[i]+acumulada[i-1])  

acumulada = np.array(acumulada)
acumulada_f = list(acumulada/sum(numeros))
acumulada_f
```

- Inventa una funcion para *samplear* de esa distribución de probabilidad acumulada y samplea otros 1000 puntos con este método.


```python
# para samplear (coger muestras) cogemos un valor de Y y ver a que X se corresponde

#Forma 1:
#sampleo1 = random.sample(acumulada, 600)

#Forma 2:
sampleo2 = []
for i in range(600):
    sampleo2.append(random.choice(acumulada_f))
```

- Compara ambas distribucion en un histograma

```python

plt.hist(acumulada_f, bins=200,label='Normal acumulada', color = "red")
plt.hist(sampleo2, bins=200, label='Sampleo DT Normal', color = "blue")

plt.legend()
plt.show()
```
