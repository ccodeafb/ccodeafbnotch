```python
import pandas as pd

orders = pd.read_csv("orders.csv")
products = pd.read_csv("products.csv")
customers = pd.read_csv("customers.csv")
```

```python
orders
```

```python
customers
```

```python
pd.merge(orders,customers)    #no recomendable
```

```python
orders.merge(customers, on="customer_id")        #si recomendable, especificando la columna en q se basa el merge
```

```python
orders.loc[8]= (9, 1, 5, 6, "2017-02-02")   ##creando una fila nueva para mostrar la diferencia entre left y right
```

```python
orders.merge(products, on="product_id", how="left")   #los nan: esta cogiendo segun la izquierda pero no lo encuentra en products (id = 5), asiq pone un nan. Todo esto por el "left"
```

```python
orders.merge(products, on="product_id", how="right") #aqui si encuentra todo, porque busca lo que hay a la derecha, q no esta el nuevo id
```

```python
sorted(orders.merge(products, on="product_id", how="left")["product_id"].unique())   #ids con left
```

```python
sorted(orders.merge(products, on="product_id", how="right")["product_id"].unique())   #ids con right
```

```python
sorted(orders.merge(products, on="product_id", how="outer")["product_id"].unique())   #ids con right + left
```

```python
sorted(orders.merge(products, on="product_id", how="inner")["product_id"].unique())   #ids en right y en left
```
