```python
import matplotlib.pyplot as plt
plt
```

```python
x = np.linspace(-5,5,40)   #creamos el array de -5 a 5 con 40 elementos (n de pasos)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(y)
```

```python
plt.xlabel("eje x")
plt.ylabel("eje y")
plt.title("Seno de -5 a 5")

plt.plot(x,y,linestyle='dashdot',marker='o')   #corregimos con "x" en este caso para q el eje x vaya desde -5 a 5
```

```python
# Generar datos aleatorios
labels = ['A', 'B', 'C', 'D']
values = np.random.randint(low=1, high=1000, size=4)

# Crear el grÃ¡fico
plt.bar(labels, values, color="olivedrab")
plt.title("GrÃ¡fico de barras")
plt.xlabel("CategorÃ­as")
plt.ylabel("Valores")
plt.show()
```

```python
f = np.random.randn(100)
g = np.random.randn(100)

plt.scatter(f, g)

plt.title("GrÃ¡fico de dispersiÃ³n")
plt.xlabel("Valores f")
plt.ylabel("Valores g")
plt.show()
```

```python
data = np.random.randn(100)

plt.hist(data, bins=5, color = "Red")

plt.title("Histograma")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.show()
```

```python
#cuando tenemos diferentes plots en una figura, los cambios de titulo, etc se realizan con set_:


fig, ax = plt.subplots(2, 2, figsize=(13, 8)) # 2 filas, 2 columnas
ax[0, 0].plot([1, 2, 3, 4], [1, 4, 2, 3]) # Dibuja en el primer subplot
ax[0, 1].bar([1, 2, 3, 4], [1, 4, 2, 3]) # Dibuja en el segundo subplot
ax[0, 1].set_title('Titulo del Bar')
ax[1, 0].scatter([1, 2, 3, 4], [1, 4, 2, 3]) # Dibuja en el tercer subplot
ax[1, 0].set_ylabel('Eje Y de la nube de puntos')
ax[1, 1].pie([10, 20, 30, 40]) # Dibuja en el cuarto subplot
```
