* Dentro de la base de datos que acabamos de crear, crea tablas para almacenar la información referente a los discos: los interpretes, grupos musicales, las discográficas y las productoras, etc.  

* Crear estas tablas usando Python. Señala los campos que serian clave primaria.

```python
import pymysql # Si no lo tienes instalado se instala con: pip install pymysql

miConexion = pymysql.connect( host='localhost', user= 'root', password='mololoco97', db='discos')

cur = miConexion.cursor()
sql = "CREATE TABLE albumes (id INT AUTO_INCREMENT PRIMARY KEY, album VARCHAR(200), n_canciones INT, artista VARCHAR(60), genero VARCHAR(60))"
sql1 = "CREATE TABLE artistas (id INT AUTO_INCREMENT PRIMARY KEY, artista VARCHAR(60), grammys INT, n_integrantes INT)"
sql2 = "CREATE TABLE generos (id INT AUTO_INCREMENT PRIMARY KEY, genero VARCHAR(60), año_origen INT, pais_origen VARCHAR(60))"
cur.execute(sql)
cur.execute(sql1)
cur.execute(sql2)
```

```python
rows = cur.fetchall()
print ("Tabla Creada")
print(rows)
```
