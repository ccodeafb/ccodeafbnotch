## Reto 2

    * El desarrollo de estas consultas se tiene que hacer tanto en Workbench como en Python
    
* Obt√©n  el  id  y  la  nota  de  los  alumnos  que  tengan  un  id  entre  1  y 20,  o  que  tenga una nota mayor  de 8 y la nota tenga fecha del a√±o pasado.



```python
try:
    cur = miConexion.cursor()
    sql = "SELECT FK_student_mark, AVG (mark) FROM marks WHERE FK_student_mark BETWEEN '1' OR '20' GROUP BY FK_student_mark HAVING AVG (mark) > 8"  
    cur.execute(sql)
    miConexion.commit()   
        
except Exception as e:
    print("Exception occured:{}".format(e))
```

```python
try:
    cur = miConexion.cursor()
    sql = "SELECT FK_student_mark, mark FROM marks WHERE FK_student_mark BETWEEN '1' OR '20' AND mark > 8"  
    cur.execute(sql)
    miConexion.commit()   
        
except Exception as e:
    print("Exception occured:{}".format(e))
```


* Obt√©n la  media de las notas que se han  dado en el √∫ltimo  a√±o por asignatura. 




```python
try:
    cur = miConexion.cursor()
    sql = "SELECT idmarks, AVG(mark) FROM marks WHERE date LIKE '2023%' GROUP BY idmarks"      
    cur.execute(sql)
    miConexion.commit()   
        
except Exception as e:
    print("Exception occured:{}".format(e))
```


* Obt√©n  la  media  aritm√©tica de las notas que se han  dado en el √∫ltimo  a√±o  por  alumno. 




```python
try:
    cur = miConexion.cursor()
    sql = "SELECT FK_student_mark, AVG(mark) FROM marks WHERE date LIKE '2023%' GROUP BY FK_student_mark"  
    cur.execute(sql)
    miConexion.commit()   
        
except Exception as e:
    print("Exception occured:{}".format(e))
```
