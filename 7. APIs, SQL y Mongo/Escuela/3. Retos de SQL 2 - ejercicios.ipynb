* Crea una BD de mysql nueva donde cargues informacin de alojamientos y barrios de Madrid de AirBnB. La informacin est disponible [aqu](http://insideairbnb.com/get-the-data.html). 

* Los ficheros a cargar son listings.csv y neighbourhoods.csv.

* Una vez cargados, prueba a hacer algunas consultas que sean de tu inters.


```python
from sqlalchemy import create_engine 
import pymysql
import pandas as pd
```

```python
miConexion = pymysql.connect( host='localhost', user= 'root', password = "administrador", db='airbnb_madrid' )
```

```python
#Leemos los csv con pandas
alojamientos = pd.read_csv('listings.csv')
barrios = pd.read_csv('neighbourhoods.csv')
```

```python
#Creamos el engine necesario

db_data = "mysql+pymysql://root:administrador@localhost:3306/airbnb_madrid?charset=utf8mb4"
engine = create_engine(db_data)
```

```python
#Cargamos alojamientos
alojamientos.to_sql('alojamientos', engine, if_exists='append', index=False)  
```

```python
#Cargamos barrios
barrios.to_sql('barrios', engine, if_exists='append', index=False)  
```

```python
sql = "SELECT * FROM barrios WHERE neighbourhood IN ('Delicias','Quintana')"
barrio_estudiado = pd.read_sql(sql,engine)
barrio_estudiado.head()
```

```python
sql = "SELECT * FROM alojamientos WHERE number_of_reviews_ltm > 210 AND price < 150"
barrio_estudiado = pd.read_sql(sql,engine)
barrio_estudiado
```
